## Data preparation

### 1.Remove duplicates

```python
# df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
# preview dataframe and we figure out that there is no duplicate data
df
```

### 2.Fill null values

```python
# first know there are null values or no
# then choose how to deal with it 
df.isnull().sum()
```

```python
# we know that one NaN value found in Caffeine (mg)  columns
# so show row which has null values 
df[df.isnull().any(axis=1)]
```

```python
# at First there are many ways to deal with null values 
# 1.Drop null values 
# 2.ask stakeholders for values: 
# and when search for Iced Brewed Coffee (With Milk & Classic Syrup) Beverage 
# with 2%milk Beverage_prep we find Caffeine (mg) = 125 
# if we Choose to work caffeine (mg) =125
df_clean = df.fillna(125)
df_clean.loc[[158]]
```

### 3.Drop unnecessary Columns

##### we can drop Vitamin A (% DV), Vitamin C (% DV), Calcium (% DV), Iron (% DV)  and other columns that we don't need it in our analysis now 

```python
dfcopy = df_clean.copy()
dfcopy.drop(['Trans Fat (g) ', 'Saturated Fat (g)', ' Sodium (mg)',' Total Carbohydrates (g) ', ' Dietary Fibre (g)', 'Vitamin A (% DV) ', 'Vitamin C (% DV)',
       'Iron (% DV) ',' Calcium (% DV) ', 'Caffeine (mg)'], axis=1, inplace=True)
dfcopy.head()
```

## Data Visualisations

### Q1.Which drink has the highest calories from the dataset?

```python
# sort data and show highest value of Calories
dfcopy.sort_values("Calories", ascending=False).head(2)
```

```python
# data visualization to see the which drink has the highest calories
plt.figure(figsize=(25, 10))
plt.title('Drink Beverage Calories')
plt.xticks(rotation=45, ha='right')
plt.bar(dfcopy['Beverage'], dfcopy['Calories'])
plt.xlabel("Beverage")
plt.ylabel("Calories")
plt.ylim(0 , 550)
plt.show()
```

#### From previous the graph, we see White Chocolate Mocha (Without Whipped Cream) has the highest calories in the dataset

### 2.Highest Sugar Drink

```python
# sort data and show highest value of Calories
dfcopy.sort_values(" Sugars (g)", ascending=False).head(2)
```

```python
# data visualization to see the which drink has the highest calories
plt.figure(figsize=(25, 10))
plt.title('Drink Beverage  Sugars (g)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0 , 80)
sns.barplot(x="Beverage", y=" Sugars (g)", data=dfcopy)
plt.show()
```

#### From the previous graph, we see Java Chip (Without Whipped Cream) (Venti Nonfat Milk and Whole Milk) has the highest Sugar in the dataset
